```python
#checking for types of rooms available
properties= df['room type'].value_counts().to_frame()
properties
```

```python
# visualizing room types using bar graph

room_types= plt.bar(properties.index, properties.loc[:,"count"])

plt.bar_label(room_types)
plt.title('Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

```python
#neighbourhood group with highest number of locations

locations=df['neighbourhood group'].value_counts().to_frame()
locations
```

```python
# visualzing the most populas locations from data

popular=plt.bar(locations.index, locations.loc[:,"count"], color='magenta' )
plt.bar_label(popular)
plt.title("Popular Locations")
plt.xlabel("Locations")
plt.ylabel("Count")
plt.show()
```

```python
#visualizing the avg prices in the nighbourhood group
avg_price=df.groupby('neighbourhood group')['price'].mean().sort_values(ascending=False).to_frame()
bars=plt.bar(avg_price.index, avg_price['price'], color='g' )
plt.bar_label(bars, fmt='%.0f', label_type='center', color='black')
plt.xticks(rotation=45, ha='right')
plt.title("Average Price in Neighbourhood Locations")
plt.xlabel("Locatoins")
plt.ylabel("Price")
plt.show()
```

```python
#visualizing Most Demanding Room Types and Avg Rating in Each Neighbourhood Group
demand = df.groupby(['neighbourhood group', 'room type'])['review rate number'].mean().sort_values(ascending=False).to_frame()
demand = demand.reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(data=demand, x='neighbourhood group', y='review rate number', hue='room type')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Review Rating')
plt.title('Most Demanding Room Types and Avg Rating in Each Neighbourhood Group')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Room Type')
plt.show()
```

```python
# visualizing avg review rate for all the verified and unconfirmed customers

verified=df.groupby('host_identity_verified')['review rate number'].mean().sort_values(ascending=False).to_frame()
review=plt.bar(verified.index, verified['review rate number'],color='teal')
plt.bar_label(review, fmt='%.2f', label_type='edge')
plt.title('avg review rate for all the verified customers')
plt.xlabel('average review rate')
plt.ylabel('host identity verified')
plt.show()
```
